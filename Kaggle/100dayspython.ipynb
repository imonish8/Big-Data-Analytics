
- Algorithm efficiency is defined by Big O.
- Time taken by the code is measured by number of operation the code is performing not the time taken for the code to run, this is because the running time for same code could vary on different machine but the number of operation. This Time Complexity.
- Space taken by the code while running to assign variable and etc. is measured by space complexity.
- Best case scenario, Average case scenario, Worst case scenario.
- We focus on worst case scenario, it gives the efficiency of the code as big O Notation.
- Technically there is no best case and Average case, we always calculate worst case.
- O(1) - constant time complexity - eg simple code of addition or multiplication or subtraction with no loop, plain addition or subtraction.
- O(N) - Linear Time complexity - eg, simple code of Loop through from 1 to n times. That is addition of 1 to 100 looping through 100 linear times.
- O(LogN) - Logarithmic Time complexity which is finding an element in sorted array.
- O(N^2) - Quadratic Time Complexity which often refers to Nested Loops. But may vary/
- O(2^N) - exponential Time Complexity which as example the Combinational Lock. Where the Power represents the number of dials. On the combination.

```python
def print_items(n):
    for i in range(n+1):
        print("Hello World", i)
        
        
print_items(10)

# this program has time complexity of O(n) as it performs n operation depending on input n.
# previous square program was performing only 1 operation single irrespective of numbers input n.
# This program will have different space complexity as well further discussed 
# when learned find the space complecity of this one.
```

```python
# droping constant in Big O.
def print_twice(n):
    for i in range(n):
        print(i)
    for j in range(n):
        print(j)

print_twice(10)

# actual time Complexity of the program is O(N + N) => O(2N) as the operation is done n times and twice.
# but we drop the constant to get the approximate time complexity as O(N)
# In practical examples the constant represents the hardware or the computer. we are not looking for faster/ slower computr
# we are intersted in the algorithms as constant represnts the computer, we elimate the constants to get algo info.
# As N approaches Infinity, constant factors are not a big deal.

```

```python
def print_nested(n):
    for i in range(n):
        for j in range(n):
            print((i,j))

print_nested(2)
```

```python
def pair_sum_sequence(n):
    total = 0
    for i in range(n):
        total = total + pair_sum(i, i+1)
    return total

def pair_sum(a,b):
    return a + b

pair_sum_sequence(3)
```

## Time Complexity when to add and when to multiply.

- If Algorithms is in the form of "do this, then when all done, do that" then you **add** the runtimes.
- If Algorithm is in the form of "do this for each time you do that", then you **Multiply** the runtimes.

## Time Comlexity measure for the code.

1. Rule no 1 : any asignment statemnts and if statemnts that executed once *regardless of the size* of the problem is **counted as O(1).**
2. Rule 2 : *Simple For loop* from 0 to N (with no internal loops) is **measured as O(N).**
3. rule 3 : a *nexted loop* for the same type takes quadratic time complexity **O(N^2).**
4. rule 4: a loop in which the *controlling parameter is divided* by two each step take logarithmic Time complexity,**O(log N)**
5. When dealing with multiple statemntss.,just add them up.
