```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [2,4,52,2]
plt.plot(x,y,marker='o', ms=10,mec='r', mfc='g')
```

```python
plt.plot(x,y,marker='o',color = 'hotpink', ms=10,mec='r', mfc='g')
```

```python
plt.plot(x,y,marker='o',color = 'hotpink', ms=10,mec='r', mfc='g',linestyle='dashed' )
plt.xlabel("Month")
plt.ylabel("Salary / mo")
plt.title("Boss")
```
