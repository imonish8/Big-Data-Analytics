```python
from plotnine.data import mpg
from plotnine.data import huron, economics


from plotnine import ggplot, aes, geom_point, geom_bar, stat_bin,geom_histogram, geom_boxplot, labs,geom_line, scale_x_timedelta, coord_flip

from plotnine import facet_grid, theme_dark
```

```python
mpg
```

```python
ggplot(mpg)+aes(x="class", y="hwy")+geom_point()
```

```python
ggplot(mpg) + aes(x="class") + geom_bar()
```

```python
ggplot(huron) + aes(x="level") + stat_bin(bins=10) + geom_bar()
```

```python
ggplot(huron) + aes(x="level") + geom_histogram(bins=10)
```

```python
(
  ggplot(huron)
  + aes(x="factor(decade)", y="level")
  + geom_boxplot()
  + labs("HURON VISULISATION")
  + geom_line()
  #+ coord_flip()

)
```

```python
huron
```

```python
(
    ggplot(economics)
    + aes(x="date", y="pop")
    + scale_x_timedelta(name="Years since 1970")
    + labs(title="Population Evolution", y="Population")
    + geom_line()
)
```

```python
ggplot(mpg) + aes(x="class") + geom_bar() + coord_flip()
```

```python

(
    ggplot(mpg)
    + facet_grid("year~class")
    + aes(x="displ", y="hwy")
    + labs(
        x="Engine Size",
        y="Miles per Gallon",
        title="Miles per Gallon for Each Year and Vehicle Class",
    )
    + geom_point()
    #+coord_flip()
    + theme_dark()

)
```
